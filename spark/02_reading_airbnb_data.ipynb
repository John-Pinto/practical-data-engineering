```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Reading Airbnb Data") \
    .getOrCreate()
```

```python
london_listings = spark.read.csv(path="./data/airbnb-london-listings.csv.gz",
                                header=True,
                                inferSchema=True,
                                sep=",",
                                quote='"',
                                escape='"',
                                multiLine=True,
                                mode="PERMISSIVE")
```

```python
for field in london_listings.schema:
    print(field)
```

```python
neighbourhoods = london_listings.select(london_listings.neighbourhood_cleansed)
neighbourhoods.show(20, truncate=False)
```

```python
spark.stop()
```
