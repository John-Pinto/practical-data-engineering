Starting a spark session

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("First Spark Application") \
    .getOrCreate()
```

```python
data = [
    {"userID": 1, "paymentAmount": 100.0, "date": "2025-01-01"},
    {"userID": 2, "paymentAmount": 150.5, "date": "2025-01-02"},
    {"userID": 3, "paymentAmount": 200.75, "date": "2025-01-03"},
    {"userID": 2, "paymentAmount": 250.99, "date": "2025-01-04"},
    {"userID": 1, "paymentAmount": 300.0, "date": "2025-01-05"}
]

df = spark.createDataFrame(data)
```

```python
df.show()
```

```python
data = [
    (1, 100.0, "2025-01-01"),
    (2, 150.5, "2025-01-02"),
    (3, 200.75, "2025-01-03"),
    (2, 250.99, "2025-01-04"),
    (1, 300.0, "2025-01-05")
]

df = spark.createDataFrame(data=data, schema=["userID", "paymentAmount", "date"])
df.show(3)
```
